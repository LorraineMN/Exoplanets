### a. Using the model tracks you downloaded, generate a diagram showing the mass-radius curves for planets with the following compositions: 100% Fe, 50% Fe, 30%Fe, Rock, 25%H2O, 100%H2O, and cold H2/He.

### b. Why aren’t these curves simply lines of constant density?

The more massive the planet, the higher gravity will be, so material should condense. So, given two planets of the same composition but with different masses, the more massive planet should always be denser.
